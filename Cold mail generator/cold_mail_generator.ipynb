```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader("https://southasiacareers.deloitte.com/job/Delhi-Adobe-Analytics-T&T-Customer-Sr_Con/38982144/")
page_data=loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPPED TEXT FROM WEBSITE:
    {page_content}
    ### INSTRUCTION:
    The scraped text is from the career page of a website.
    Your job is to extract the job postings and return them in JSON format containing the following keys:
    'role', 'experience', 'skills' and 'description'.
    Only return the valid JSON.
    ### VALID JSON (NO PREAMBLE):
    """
)

chain_extract = prompt_extract | llm
res = chain_extract.invoke({'page_content': page_data})
print(res.content)
```

```python
type(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser
json_parser=JsonOutputParser()
json_res=json_parser.parse(res.content)
json_res
```

## Prepare the chromadb

Add all the portfolio links and relevant skills in the database.

```python
import pandas as pd
df=pd.read_csv("my_portfolio.csv")
df.head()
```

```python
pip install chromadb
```

```python
import chromadb
client=chromadb.PersistentClient('vectorstore')       #Client - create chromadb in the memory, PersistentClient - in the disk
collection=client.get_or_create_collection(name='portfolio')
```

```python
import uuid
```

```python
if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents=row["Techstack"],
                       metadatas={"links": row["Links"]},
                       ids=[str(uuid.uuid4())])
```

```python
links=collection.query(query_texts=["Experience in Python","Expertise in react"],
                       n_results=2).get('metadatas')

links
```

```python
prompt_email = PromptTemplate.from_template(
        """
        ### JOB DESCRIPTION:
        {job_description}

        ### INSTRUCTION:
        You are Akankshya, a business development executive at AtliQ. AtliQ is an AI & Software Consulting company dedicated to facilitating
        the seamless integration of business processes through automated tools.
        Over our experience, we have empowered numerous enterprises with tailored solutions, fostering scalability,
        process optimization, cost reduction, and heightened overall efficiency.
        Your job is to write a cold email to the client regarding the job mentioned above describing the capability of AtliQ
        in fulfilling their needs.
        Also add the most relevant ones from the following links to showcase Atliq's portfolio: {link_list}
        Remember you are Akankshya, BDE at AtliQ.
        Do not provide a preamble and do write in a organized format, so tht I do not need to scroll to the right.
        ### EMAIL (NO PREAMBLE):

        """
        )

chain_email = prompt_email | llm
res = chain_email.invoke({"job_description": str(job), "link_list": links})
print(res.content)
```
